```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# from lmfit import minimize, Parameters, report_fit
```

1. 300g grating PL

```python
pl_data = pd.read_csv(r'D:\LightFieldData\20230606_sic_epitaxy_Vsi\\20230606_sic_epitaxy_Vsi_300g_IntTime 1s_Temp5.58K_Power6.82mW_exportedGrapgh_PL',delimiter=r"\s+|\t+")
wavelength_data = pl_data['Wavelength(nm)']
counts_data = pl_data['Intensity']
x = []
y = []
for item in wavelength_data:
    x.append(item)
for item in counts_data:
    y.append(item)    
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('Wavelength(nm)')
ax.set_ylabel('Intensity')
ax.text(910,12000,'300g-grating',color='r',fontsize=16)
```

2. Power Mapping 1200g

```python
pl_data = pd.read_csv(r'D:\LightFieldData\20230606_sic_epitaxy_Vsi\\20230606_sic_epitaxy_Vsi_1200g_IntTime1s_Temp5.58K_Power_4.16uWto6.82mW#PowerMapping (deg.)_71.4_116.4_exportedGraph',delimiter=r"\s+|\t+")
pl_data
deg_data = pl_data['Powermapping(deg.)']
counts_data = pl_data['Amplitude']
x = []
y = []
for item in deg_data:
    x.append(item)
for item in counts_data:
    y.append(item) 

deg_real = np.linspace(np.pi/2,0,len(x))   
p_max = 6.82
p_min = 4.16*10**(-3)
power_real = p_min + (p_max - p_min)*np.cos(deg_real)

fig, ax = plt.subplots()
ax.plot(power_real,y)
ax.set_xlabel('Power(mW)')
ax.set_ylabel('Intensity')
ax.text(1,1750,'1200g-grating',color='r',fontsize=16)
```
